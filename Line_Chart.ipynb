## Loading Libraries and Data

```r
library(tidyverse)
```

## A quick look at the data

```r
data("Orange")
```

```r
view(Orange)
```

```r
names(Orange)
```

## Basic Line Chart

```r
Orange %>%
  ggplot(aes(x = age,
             y = circumference)) +
  geom_point()
```

## Adding Color by Category

```r
Orange %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point()
```

## Customizing Points with Size and Transparency

```r
Orange %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5)
```

## Adding Lines to the Plot

```r
Orange %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5) +
  geom_line(size = 1)
```

## Enhancing the Plot with Theme and Labels

```r
Orange %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5) +
  geom_line(size = 1) +
  theme_test() +
  labs(title = "Circumference as a Function of Age",
       x = "Tree Age",
       y = "Tree Circumference")
```

## Reordering Factor Levels

```r
Orange %>%
  mutate(Tree = factor(Tree, levels = c("1", "2", "3", "4", "5"))) %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5) +
  geom_line(size = 1) +
  theme_test() +
  labs(title = "Circumference as a Function of Age",
       x = "Tree Age",
       y = "Tree Circumference")
```

## Filtering Specific Trees

```r
Orange %>%
  mutate(Tree = factor(Tree, levels = c("1", "2", "3", "4", "5"))) %>%
  filter(Tree %in% c(1, 2, 5)) %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5) +
  geom_line(size = 1) +
  theme_test() +
  labs(title = "Circumference as a Function of Age",
       x = "Tree Age",
       y = "Tree Circumference")
```

## Excluding Specific Trees

```r
Orange %>%
  mutate(Tree = factor(Tree, levels = c("1", "2", "3", "4", "5"))) %>%
  filter(Tree != 4) %>%
  ggplot(aes(age,
             circumference,
             color = Tree)) +
  geom_point(size = 4, alpha = 0.5) +
  geom_line(size = 1) +
  labs(title = "Circumference as a Function of Age",
       x = "Tree Age",
       y = "Tree Circumference") +
   theme_test()
```
